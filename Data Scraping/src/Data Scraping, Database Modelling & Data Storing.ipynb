```python
import pycountry

# Function untuk convert kode negara menjadi nama negara
def convert_to_country_name(country_code):
    try:
        country = pycountry.countries.get(alpha_2=country_code)
        return country.name
    except LookupError:
        return 'Unknown'

# Convert kode negara menjadi nama negara
Player_Basic['Nation'] = Player_Basic['Nation'].map(lambda x: convert_to_country_name(x.split()[0]))
```

### Tabel Player_Wage

```python
# Melakukan join kedua tabel
Player_Wage = pd.merge(Player_Basic,wages,left_on='PlayerName',
                       right_on='Player',how='left')

# Mem-filter tabel Player_Wage
Player_Wage = Player_Wage[['PlayerID','Weekly Wages','Annual Wages']]
```

```python
# Mengecek missing values
Player_Wage.isnull().sum()
```

Terdapat missing values bukan berarti pemain tersebut tidak digaji, melainkan tidak terdapat data gaji pemain tersebut. Oleh karena itu, kita dapat membiarkan missing values tersebut.

```python
Player_Wage.head()
```

Untuk mempermudah kalkulasi di database, kita perlu mengubah nilai dari wages menjadi integer. Kita akan mengambil gaji dalam mata uang euro (angka di luar tanda kurung).

```python
import re

def extract_number_from_string(string):
    match = re.search(r'(\d+(?:,\d+)?)', string)  # Mencari angka di luar kurung
    if match:
        number = match.group(1)  # Extract angka tersebut
        number = number.replace(',', '')  # Menghilangkan koma
        return int(number)  # Convert ke integer
    else:
        return None
```

```python
# Apply fungsi tersebut untuk convert ke integer
Player_Wage['Weekly Wages'] = Player_Wage['Weekly Wages'].astype('str').apply(extract_number_from_string)
Player_Wage['Annual Wages'] = Player_Wage['Annual Wages'].astype('str').apply(extract_number_from_string)
Player_Wage['Annual Wages'] = Player_Wage['Annual Wages'] * 1000
```

```python
# Mengubah nama column 
Player_Wage.rename(columns={'Weekly Wages': 'WeeklyWages', 
                            'Annual Wages': 'AnnualWages'},
                   inplace=True)
```

```python
# Create Referee table
cursor.execute('DROP TABLE IF EXISTS referee')

cursor.execute('''
    CREATE TABLE referee (
        RefereeID INTEGER PRIMARY KEY,
        RefereeName TEXT,
        Nation TEXT,
        YellowCard INTEGER,
        RedCard INTEGER
    )
''')
```

```python
# Create Player_Wage table
cursor.execute('DROP TABLE IF EXISTS stadium')

cursor.execute('''
    CREATE TABLE stadium (
        StadiumID INTEGER PRIMARY KEY,
        Team TEXT,
        Capacity INTEGER,
        Street TEXT,
        City TEXT,
        Country TEXT
    )
''')
```

```python
# Create Matches table
cursor.execute('DROP TABLE IF EXISTS matches')

cursor.execute('''
    CREATE TABLE matches (
        MatchID INTEGER PRIMARY KEY,
        Date DATE,
        Competition TEXT,
        Venue TEXT,
        GoalsFor INTEGER,
        GoalsAgainst INTEGER,
        Opponent TEXT,
        RefereeID INTEGER,
        StadiumID INTEGER,
        FOREIGN KEY (RefereeID) REFERENCES referee(RefereeID),
        FOREIGN KEY (StadiumID) REFERENCES stadium(StadiumID)
    )
''')
```

```python
# Create Match_Goal_Logs table
cursor.execute('DROP TABLE IF EXISTS match_goal_logs')

cursor.execute('''
    CREATE TABLE match_goal_logs (
        MatchID INTEGER,
        GoalOrder INTEGER,
        ScorerPlayerID TEXT,
        AssistPlayerID TEXT,
        PRIMARY KEY (MatchID, GoalOrder),
        FOREIGN KEY (MatchID) REFERENCES matches(MatchID)
    )
''')
```

```python
# Create Player_Basic table
cursor.execute('DROP TABLE IF EXISTS player_basic')

cursor.execute('''
    CREATE TABLE player_basic (
        PlayerID INTEGER PRIMARY KEY,
        PlayerName TEXT,
        Nation TEXT,
        Position TEXT,
        Age INTEGER
    )
''')
```

```python
# Create Player_PlayingTime table
cursor.execute('DROP TABLE IF EXISTS player_playingtime')

cursor.execute('''
    CREATE TABLE player_playingtime (
        PlayerID INTEGER PRIMARY KEY,
        MatchPlayed INTEGER,
        Starts INTEGER,
        Minutes REAL,
        90sPlayed REAL,
        FOREIGN KEY (PlayerID) REFERENCES player_basic(PlayerID)
    )
''')
```

```python
# Create Player_Performance table
cursor.execute('DROP TABLE IF EXISTS player_performance')

cursor.execute('''
    CREATE TABLE player_performance (
        PlayerID INTEGER PRIMARY KEY,
        Goals INTEGER,
        Assist INTEGER,
        GoalAssist INTEGER AS (Goals + Assist),
        FOREIGN KEY (PlayerID) REFERENCES player_basic(PlayerID)
    )
''')
```

```python
# Create Goalkeeper_Performance table
cursor.execute('DROP TABLE IF EXISTS goalkeeper_performance')

cursor.execute('''
    CREATE TABLE goalkeeper_performance (
        PlayerID INTEGER PRIMARY KEY,
        GoalsAgainst REAL,
        GoalsAgainst90 REAL,
        SavePct REAL,
        CleanSheetPct REAL,
        FOREIGN KEY (PlayerID) REFERENCES player_basic(PlayerID)
    )
''')
```

```python
# Create Player_Wage table
cursor.execute('DROP TABLE IF EXISTS player_wage')

cursor.execute('''
    CREATE TABLE player_wage (
        PlayerID INTEGER PRIMARY KEY,
        WeeklyWages REAL,
        AnnualWages REAL,
        FOREIGN KEY (PlayerID) REFERENCES player_basic(PlayerID)
    )
''')
```
