```python
import pandas as pd
import numpy as np
df = pd.read_csv('raw_data.csv')
```

```python
df.head()
```

```python
def splitting(s):
  s['hotel_name'] = s['hotel_name'].split('.')[1]
  return s

def splitting2(s):
  s['address'] = s['address'].split('/"')[0]
  return s

df = df[df.hotel_name != 'None']
df = df[df.review != '0.0']
df = df[df.address != 'None']
df = df[df.prices != 0]
df.drop(df.columns[df.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
df = df.apply(splitting , axis = 1)
df = df.apply(splitting2 , axis = 1)
print(df)
df.to_csv('data_cleansed.csv', index=False)
```
