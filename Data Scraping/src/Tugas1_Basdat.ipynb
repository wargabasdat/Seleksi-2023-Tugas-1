```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://books.toscrape.com/catalogue/category/books/travel_2/index.html"  # Ganti dengan URL yang sesuai
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

produk = soup.find_all("li", class_="col-xs-6 col-sm-4 col-md-3 col-lg-3")
i = 1
nama_buku = []
harga_buku = []
stock_buku = []
for p in produk:
    nama = p.find('h3')
    if nama:
        nama_buku.append(nama.text.strip())
    else:
        print("Product title not found")

    harga = p.find('p', class_='price_color')
    if harga:
        harga_buku.append(harga.text.strip())
    else:
        print("Price not found")

    avail = p.find('p', class_='instock availability')
    if avail:
        stock_buku.append(avail.text.strip().replace('\n', '').replace('    ', ''))
    else:
        print("Out of Stock")

produk_dict = {'nama buku': nama_buku, 'jenis': "Travel", 'harga': harga_buku, 'stock': stock_buku}
df = pd.DataFrame(produk_dict, columns=['nama buku', 'jenis', 'harga', 'stock'])

print(df)



    
```

```python
#Misteri book
import pandas as pd
import requests
from bs4 import BeautifulSoup

url = "https://books.toscrape.com/catalogue/category/books/mystery_3/index.html"  # Ganti dengan URL yang sesuai
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

produk = soup.find_all("li", class_="col-xs-6 col-sm-4 col-md-3 col-lg-3")
i = 1
nama_bukuMisteri = []
harga_bukuMisteri = []
stock_bukuMisteri = []
for p in produk:
    nama = p.find('h3')
    if nama:
        nama_bukuMisteri.append(nama.text.strip())
    else:
        print("Product title not found")
        
    harga = p.find('p', class_='price_color')
    if harga:
        harga_bukuMisteri.append(harga.text.strip())
    else:
        print("Price not found")
        
    avail = p.find('p', class_='instock availability')
    if avail:
        stock_bukuMisteri.append(avail.text.strip().replace('\n', '').replace('    ', ''))
    else:
        print("Out of Stock")

produk_dict = {
    'nama buku': nama_buku + nama_bukuMisteri,
    'jenis': ['Travel'] * len(nama_buku) + ['Misteri'] * len(nama_bukuMisteri),
    'harga': harga_buku + harga_bukuMisteri,
    'stock': stock_buku + stock_bukuMisteri
}

df = pd.DataFrame(produk_dict, columns=['nama buku', 'jenis', 'harga', 'stock'])
print(df)
```
