```python
# Create an instance of the Firefox web driver
driver = webdriver.Chrome()

# Navigate to the desired website
data_list = []
index = 0
for i in range(0, 75):
    url = 'https://www.sociolla.com/139-makeup?page=' + str(i)
    driver.get(url)

    # Wait until the desired elements are present
    wait = WebDriverWait(driver, 40, ignored_exceptions = (NoSuchElementException, StaleElementReferenceException))  # Wait for a maximum of 40 seconds

    #products_container = wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'product-list')))
    products = wait.until(EC.visibility_of_all_elements_located((By.CLASS_NAME, 'loaded-item')))
    #product_info = wait.until(EC.visibility_of_all_elements_located((By.CLASS_NAME, 'product__info')))
    
    for row in products:
        try:
            brand = row.find_element(By.CLASS_NAME, 'product__brand').get_attribute('innerHTML')
            name = row.find_element(By.CLASS_NAME, 'product__name').find_element(By.TAG_NAME, 'a').text
            price = row.find_element(By.CLASS_NAME, 'product__price').get_attribute('innerHTML')
            review = row.find_element(By.CLASS_NAME, 'product__reviews').get_attribute('innerHTML')

            data_list.append({
                'index': index,
                # 'discount': disc,
                'brand name': brand,
                'product name': name,
                'price': price,
                # 'more attribute': attr,
                'review number': review
            })
            index += 1  # Increment the index
        except:
            pass
    time.sleep(5)

# Close the web driver
driver.quit()
```

```python
from selenium.common.exceptions import TimeoutException

# Create an instance of the Chrome web driver
driver = webdriver.Chrome()
wait = WebDriverWait(driver, 60)
brct_data = []

# Iterate over each country element
for i in range(0, 19):
    # Navigate to the desired website
    url = country_urls[i]
    driver.get(url)

    # Retry logic
    retries = 3  # Number of retries
    while retries > 0:
        try:
            # Wait for the products to load
            wait.until(EC.visibility_of_element_located((By.XPATH, '//*[@id="sidebar"]/div[2]/div')))
            break  # Break out of the loop if successful
        except TimeoutException:
            retries -= 1
            if retries == 0:
                print(f"Failed to load products for {url}")
                break
            else:
                print(f"Retrying... ({retries} retries left)")
                continue

    # Scrape the brand names from the products
    brand_elements = driver.find_elements(By.TAG_NAME, 'label')
    brand_names = [brand.text for brand in brand_elements]

    # Print the brand names
    for brand in brand_names:
        brct_data.append({
            'country name': df_bo['brand_origins'].iloc[i],
            'brand name': brand
        })
    time.sleep(5)

# Close the web driver
driver.quit()
```

```python
df_brct = pd.DataFrame(brct_data)

# Define the values to be excluded
excluded_values = ["SALE", "All Ratings", "5.0", "4.0", "3.0", "2.0", "1.0", ""]

# Filter the DataFrame by the mask
brct_filtered = df_brct[~df_brct.iloc[:, 1].isin(excluded_values)]
brct_filtered.reset_index(drop=True, inplace=True)
brct_filtered
```

```python
# Make a dictionary
data = {
    'products': json.loads(df.to_json(orient='records')),
    'best_seller': json.loads(df_bs.to_json(orient='records')),
    'just_arrived': json.loads(df_ja.to_json(orient='records')),
    'brand_list': json.loads(df_bl.to_json(orient='records')),
    'brand_origins': json.loads(df_bo.to_json(orient='records')),
    'brand_country': json.loads(brct_filtered.to_json(orient='records'))
}

# Convert the dictionary to a JSON string
json_str = json.dumps(data, indent=2)

# Write the JSON string to a file
with open('sociolla.json', 'w') as file:
    file.write(json_str)
```
