```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
driver.get('https://liquipedia.net/valorant/VCT/2023/Pacific_League/Statistics')
```

```python
playerNames = driver.find_elements(By.XPATH,'//tbody/tr/td[1]/a')

for playerName in playerNames:
    print(playerName.text)
```

```python
countries = driver.find_elements(By.XPATH,'//tbody/tr/td[1]/span/a/img')
countryValue = []
for country in countries:
    v = country.get_attribute('alt')
    countryValue.append(v)
print(countryValue)
```

```python
len(countryValue)
```

```python
teams = driver.find_elements(By.XPATH,'//tbody/tr/td[2]/span/span[1]/a')
teamsValue = []
for team in teams:
    t = team.get_attribute('title')
    teamsValue.append(t)
print(teamsValue)
```

```python
agentsValue = []


for i in range(1,len(countryValue)+1):
    xpath = '//tbody/tr[' + str(i) + ']/td[3]/a'
    agents = driver.find_elements(By.XPATH,xpath)
    temp = []
    for agent in agents:
        ag = agent.get_attribute('title')
        temp.append(ag)
    agentsValue.append(temp)

print(agentsValue)
```

```python
mapcount = driver.find_elements(By.XPATH,'//tbody/tr/td[4]')
kill = driver.find_elements(By.XPATH,'//tbody/tr/td[5]')
death = driver.find_elements(By.XPATH,'//tbody/tr/td[6]')
assist = driver.find_elements(By.XPATH,'//tbody/tr/td[7]')
kd = driver.find_elements(By.XPATH,'//tbody/tr/td[8]')
kda = driver.find_elements(By.XPATH,'//tbody/tr/td[9]')
acs = driver.find_elements(By.XPATH,'//tbody/tr/td[10]')
kpermap= driver.find_elements(By.XPATH,'//tbody/tr/td[11]')
dpermap = driver.find_elements(By.XPATH,'//tbody/tr/td[12]')
apermap = driver.find_elements(By.XPATH,'//tbody/tr/td[13]')
```

```python
result = []

for i in range (len(countryValue)):
    temp_result = { 'Player Name' : playerNames[i].text,
                   'Country' : countryValue[i],
                   'Team' : teamsValue[i],
                   'Agents' : agentsValue[i],
                   'Maps' : mapcount[i].text,
                   'Kills' : kill[i].text,
                   'Deaths' : death[i].text,
                   'Assists' : assist[i].text,
                   'KD' : kd[i].text,
                   'KDA' : kda[i].text,
                   'ACS/Map' : acs[i].text,
                   'K/Map' : kpermap[i].text,
                   'D/Map' : dpermap[i].text,
                   'A/Map' : apermap[i].text,
                   }
    result.append(temp_result)
```

```python
df_data = pd.DataFrame(result)
df_data
```

```python
json_path = '../data/playerstats.json'
df_data.to_json(json_path)
```
