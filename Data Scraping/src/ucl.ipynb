# DATA PREPROCESSING

```python
Results = ucl_results.copy()

Results['ResultID'] = Results.index + 1
Results = Results[['ResultID', 'Round', 'Date', 'Home', 'Score', 'Away', 'Venue', 'Referee']]
Results.dropna(subset=['Round'], inplace=True)
```

```python
Results['Home'] = Results['Home'].str.rsplit(' ',1).str[0]
Results['Away'] = Results['Away'].str.split(' ',1).str[1]
Results = Results[Results['Date'] != 'Date']
```

```python
Results
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
Referee = pd.DataFrame(Results['Referee'].unique(),columns=['RefereeName'])
Referee['RefereeID'] = label_encoder.fit_transform(Referee['RefereeName'])

Results['Referee'] = label_encoder.transform(Results['Referee'])
Results = Results.rename(columns={'Referee' : 'RefereeID'})
Referee['Age'] = [30,23,56,43,53,34,65,54,34,12,43,65,32,54,23,34,66,43,32,45,32,34,54,23,34,53,34,23,54,23,43,12,44,32,14,54,23,53,44,34,54]
Referee['Nation'] = 'Nation'
Referee = Referee [['RefereeID', 'RefereeName','Age','Nation']]
Referee = Referee[Referee['RefereeName'] != 'Referee']
```

```python
Referee
```

```python
Stadium = pd.DataFrame(Results['Venue'].unique(), columns=['StadiumName'])
Stadium['StadiumID'] = label_encoder.fit_transform(Stadium['StadiumName'])

Results['Venue'] = label_encoder.transform(Results['Venue'])
Results = Results.rename(columns={'Venue' : 'StadiumID'})
Stadium['Country'] = ['Croatia','Germany','UK','Portugal','USA','Spain','France','Netherlands','Netherlands','UK','Italy','Spain','Spain','Italy','Belgium','Portugal','Czechia','UK','Portugal','Germany','France','Germany','Poland','UK','UK','UK','Denmark','Germany','Spain','Israel','Portugal','Turkey']
Stadium = Stadium[['StadiumID', 'StadiumName','Country']]
Stadium = Stadium[Stadium['StadiumName'] != 'Venue']
```

```python
Teams['Nation'] = Teams['TeamName'].map(lambda x : x.split()[0])
Teams['TeamName'] = Teams['TeamName'].str.split(' ',1).str[1]
Teams = Teams[['TeamName', 'Nation','NumPlayers','MatchPlayed','Goals','Assists']]
Teams
```

```python
import pycountry

def convert_country_code(code):
    try:
        if len(code)==2 :
            country = pycountry.countries.get(alpha_2 = code)
        elif len(code) == 3:
            country = pycountry.countries.get(alpha_3 = code)
        return country.name
    except AttributeError:
        return 'England'
    
#Teams['Nation'] = Teams['TeamName'].map(lambda x : convert_to_country_name(x.split()[0]))
Teams['Nation'] = Teams['Nation'].apply(convert_country_code)
```

# Database Modeling

```python
import mysql.connector

con = mysql.connector.connect(
    host='localhost',
    user = 'root',
    password = '******',
    database = 'championsleague'
)

interact = con.cursor()
```

```python
interact.execute('DROP TABLE IF EXISTS teams')

interact.execute('''
    CREATE TABLE teams (
        TeamName VARCHAR(100) PRIMARY KEY,
        Nation VARCHAR(100),
        NumPlayers INTEGER,
        MatchPlayed INTEGER,
        Goals INTEGER,
        Assists INTEGER
    )
''')
```

```python
interact.execute('DROP TABLE IF EXISTS stadium')

interact.execute('''
    CREATE TABLE stadium (
        StadiumID INTEGER PRIMARY KEY,
        StadiumName VARCHAR(100),
        Country VARCHAR(100)
    )
''')
```

```python
interact.execute('DROP TABLE IF EXISTS referee')

interact.execute('''
    CREATE TABLE referee (
        RefereeID INTEGER PRIMARY KEY,
        RefereeName VARCHAR(100),
        Age INTEGER,
        Nation VARCHAR(100)
    )
''')
```

```python
interact.execute('DROP TABLE IF EXISTS results')

interact.execute('''
    CREATE TABLE results (
        ResultID INTEGER PRIMARY KEY,
        Round VARCHAR(100),
        Date DATE,
        Home VARCHAR(100),
        Score VARCHAR(100),
        Away VARCHAR(100),
        StadiumID INTEGER,
        RefereeID INTEGER,
        FOREIGN KEY (Home) REFERENCES teams(TeamName),
        FOREIGN KEY (Away) REFERENCES teams(TeamName),
        FOREIGN KEY (StadiumID) REFERENCES stadium(StadiumID),
        FOREIGN KEY (RefereeID) REFERENCES referee(RefereeID)
    )
''')
```

```python
Teams.to_sql('teams', engine, if_exists='append', index=False)
Stadium.to_sql('stadium', engine, if_exists='append', index=False)
Referee.to_sql('referee', engine, if_exists='append', index=False)
Results.to_sql('results', engine, if_exists='append', index=False)
```
