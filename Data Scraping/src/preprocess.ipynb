```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# import the json file
dir = os.getcwd()
df = pd.read_json(dir[:-3]+"data\data.json", orient="index")
df.head()
```

```python
# Seperate currency 

priceOriCurrArr = list()
priceOriValArr = list()
shipAvailabilityArr = list()
shipPriceCurrArr = list()
shipPriceValArr = list()

for i in range(len(df)):
    # Make sure listing_id is an int
    df.iloc[i,0] = int(df.iloc[i,0])

    # Seperate price_original
    price_ori = df.iloc[i, 5].replace(",","")
    alreadyFound = False
    lastIdx = len(price_ori)
    firstIdx = 0
    for j in range(len(price_ori)):        
        if (price_ori[j].isnumeric()):
            if (not alreadyFound):
                firstIdx = j
                alreadyFound = True
        else:
            if (alreadyFound and price_ori[j] != "."):
                lastIdx = j
                alreadyFound = False

    priceOriCurrArr.append(price_ori[:firstIdx])
    priceOriValArr.append(float(price_ori[firstIdx:lastIdx]))

    # Make sure sold is an int
    sold_attrib = df.iloc[i,4]
    if (sold_attrib is not None):
        df.iloc[i,4] = int(sold_attrib.replace(",",""))

    # Delete currency in price_idr
    df.iloc[i,6] = float(df.iloc[i,6][3:].replace(",",""))

    # Seperate for shipping_price
    ship_price = df.iloc[i,7]
    if (ship_price == None):
        shipAvailabilityArr.append(False)
        shipPriceCurrArr.append(None)
        shipPriceValArr.append(None)
    elif (ship_price == "Free"):
        shipAvailabilityArr.append(True)
        shipPriceCurrArr.append(None)
        shipPriceValArr.append(0.0)
    else :
        shipAvailabilityArr.append(True)
        alreadyFound = False
        lastIdx = len(ship_price)
        firstIdx = 0
        ship_price = ship_price.replace(",","")
        for j in range(len(ship_price)):
            if (ship_price[j].isnumeric()):
                if (not alreadyFound):
                    firstIdx = j
                    alreadyFound = True
            else:
                if (alreadyFound and ship_price[j] != "."):
                    lastIdx = j
                    alreadyFound = False

        shipPriceCurrArr.append(ship_price[:firstIdx])
        shipPriceValArr.append(float(ship_price[firstIdx:lastIdx]))

df['price_original_currency'] = priceOriCurrArr
df['price_original_value'] = priceOriValArr
df['shipping_availability'] = shipAvailabilityArr
df['shipping_price_currency'] = shipPriceCurrArr
df['shipping_price_value'] = shipPriceValArr
df.head()
```

```python
# Drop unused attributes
df = df.drop(['price_original', 'shipping_price_original'],axis = 1)
df.head()
```

```python
# Erase "available" in available and "returns" in returns

for i in range(len(df)):
    if (df.iloc[i,3] is not None):
        df.iloc[i,3] = df.iloc[i,3].replace("available", "")
    if (df.iloc[i,7]  is not None):
        df.iloc[i,7] = df.iloc[i,7].replace("returns", "")

df.head()
```

```python
# Drop duplicates product
df = df.drop_duplicates(subset="listing_id")
df.head()
```

```python
# Split the original dataframe into four dfs
product = df[['listing_id','title', 'type', 'available', 'sold', 'return', 'seller', 'condition']]
price = df[['price_original_currency', 'price_original_value', 'price_idr']]
shipment = df[['shipping_availability', 'shipping_price_currency', 'shipping_price_value', 'shipping_location']]
payment = df[['listing_id','payments']]
```

```python
# Split payments value arr to each one element
new_payment = list()
for i in range(len(payment)):
    tempArr = payment.iloc[i,1].split(",")
    for j in tempArr:
        new_payment.append( (payment.iloc[i,0], j) )

numpy_payment = np.array(new_payment)
df_payment = pd.DataFrame(numpy_payment)
```

```python
dir = os.getcwd()
product.to_json(dir[:-3]+"data\data_product.json", orient='index', indent=2)
price.to_json(dir[:-3]+"data\data_price.json", orient='index', indent=2)
shipment.to_json(dir[:-3]+"data\data_shipment.json", orient='index', indent=2)
df_payment.to_json(dir[:-3]+"data\data_payment.json", orient='index', indent=2)
```
