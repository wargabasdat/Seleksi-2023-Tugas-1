```python
import matplotlib.pyplot as plt

# Membaca file CSV
dataframe = pd.read_csv('pubgdatabase.csv')

# Membuat scatter plot antara "rank_total" dan "damage"
plt.scatter(dataframe['rank_total'], dataframe['damage'])
plt.xlabel('Rank Total')
plt.ylabel('Damage')
plt.title('Scatter Plot: Rank Total vs Damage')

# Menampilkan plot
plt.show()

# Membuat scatter plot antara "rank_total" dan "wwcd"
plt.scatter(dataframe['rank_total'], dataframe['wwcd'])
plt.xlabel('Rank Total')
plt.ylabel('WWCD')
plt.title('Scatter Plot: Rank Total vs WWCD')

# Menampilkan plot
plt.show()
```
